```python
from pandas_datareader import data
import datetime as dt
```

## data reader

```python
start = dt.datetime(2020, 1, 2)
end = dt.datetime.today()

price = data.DataReader('GOOG', 'yahoo', start, end)
price.head()
```

```python
price['Volume']
```

```python
price['Volume'].plot(figsize = (15,8))

plt.grid()
plt.legend(fontsize = 12)
plt.xlabel('Date', fontsize = 12)
plt.ylabel('Volume', fontsize = 12)
plt.title('Volume timeline of Google')
plt.show()
```

```python
price[['Open', 'Close']].plot(figsize = (15, 8)) 

plt.grid()
plt.legend(fontsize = 12)
plt.xlabel('Date', fontsize = 12)
plt.ylabel('Price', fontsize = 12)
plt.title('Close and Open Price timeline of Google')
#plt.xlim(('2020-05-07', '2021-05-07'))
plt.show()
```

```python
price.loc['2020-02-01': '2021-06-03', ['Open', 'Close']]
```

```python
price.loc['2020-02-01': '2021-06-03', ['Open', 'Close']].plot(figsize = (15, 8)) 

plt.grid()
plt.legend(fontsize = 12)
plt.xlabel('Date', fontsize = 12)
plt.ylabel('Price', fontsize = 12)
plt.title('Close and Open Price timeline of Google')
#plt.xlim(('2020-05-07', '2021-05-07'))
plt.show()
```

```python
price_2 = data.DataReader(['GOOG', 'FB'], 'yahoo', start, end)
```

```python
price_2.head()
```

```python
price_2.columns
```

```python
price_2['Open']
```

```python
price_2['Open']['GOOG']
```

```python
close_1 = price_2['Adj Close']
close_1.head()
```

```python
close_1['GOOG'].pct_change()
```

```python
percentage = close_1.pct_change()
percentage.head()
```

```python
percentage.plot(figsize = (15,8))

plt.grid()
plt.legend(fontsize = 12)
plt.xlabel('Date', fontsize = 12)
plt.ylabel('Percentage', fontsize = 12)
plt.title('Google and Facebook Percentage in price Since 2020')
#plt.xlim(('2020-05-07', '2021-05-07'))
plt.show()
```

```python
price_2['Adj Close']['GOOG'].pct_change().plot(figsize = (15,8), c = 'g')

plt.grid()
plt.legend(fontsize = 12)
plt.xlabel('Date', fontsize = 12)
plt.ylabel('Percentage', fontsize = 12)
plt.title('Google Percentage in Close price Since 2020')
#plt.xlim(('2020-05-07', '2021-05-07'))
plt.show()
```

```python
close_1
```

```python
difference = close_1.diff()
difference.head()
```

```python
difference.plot(figsize = (15, 8))

plt.grid()
plt.legend(fontsize = 12)
plt.xlabel('Date', fontsize = 12)
plt.ylabel('Difference', fontsize = 12)
plt.title('Google and Facebook Difference in Close price Since 2020')
#plt.xlim(('2020-05-07', '2021-05-07'))
plt.show()
```

# World Bank

### Imports

```python
# pip install pandas-datareader
```

```python
from pandas_datareader import wb
```

### First data

```python
gdp = wb.download(indicator= 'NY.GDP.MKTP.PP.CD', country = 'all', start = 1960, end = 2020)
```

```python
gdp.head()
```

```python
gdp.unstack().head()
```

```python
gdp2 = gdp.unstack()
gdp2.head()
```

```python
#gdp2.columns
```

```python
gdp2.columns = np.arange(1960, 2021)
gdp2.head()
```

```python
gdp2.columns
```

```python
gdp3 = gdp2.T
gdp3.head()
```

```python
gdp3.index.rename('year', inplace=True)
gdp3.head()
```

```python
gdp3['Iran, Islamic Rep.'].plot(figsize = (15,8), lw = 3)

plt.grid()
plt.legend(fontsize = 12)
plt.xlabel('year', fontsize = 12)
plt.ylabel('GDP PPP', fontsize = 12)
plt.title('GDP PPP\nSInce 1960 - 2020')
plt.show()
```

```python
gdp3[['Iran, Islamic Rep.', 'Norway', 'Canada', 'Italy']].plot(figsize = (15,8), lw = 3)

plt.grid()
plt.legend(fontsize = 12)
plt.xlabel('year', fontsize = 12)
plt.ylabel('GDP PPP', fontsize = 12)
plt.title('GDP PPP\nSInce 1960 - 2020')
plt.show()
```

### Second data

```python
gini = wb.download(indicator='SI.POV.GINI', country = 'all', start=1960, end=2020)
```

```python
gini.head()
```

```python
gini2 = gini.unstack()
gini2.head()
```

```python
gini2.columns = np.arange(1960, 2021)
gini2.head()
```

```python
gini3 = gini2.T
gini3.head()
```

```python
gini3['Iran, Islamic Rep.'].plot(figsize = (15,8), lw = 3)

plt.grid()
plt.legend(fontsize = 12)
plt.xlabel('year', fontsize = 12)
plt.ylabel('Gini index', fontsize = 12)
plt.title('Gini index\nSInce 1960 - 2020')
plt.show()
```

```python
gini3[['Iran, Islamic Rep.', 'Norway', 'Canada', 'Italy']].plot(figsize = (15,8), lw = 3)

plt.grid()
plt.legend(fontsize = 12)
plt.xlabel('year', fontsize = 12)
plt.ylabel('Gini index', fontsize = 12)
plt.title('Gini index\nSInce 1960 - 2020')
plt.show()
```
